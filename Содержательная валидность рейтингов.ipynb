# Содержательная валидность международных гендерных индексов

```python
import pandas as pd # для обработки и анализа структурированных данных
import matplotlib.pyplot as plt # для визуализации и графического представления
```

```python
# загружаем таблицу
df = pd.read_excel('construct_validity.xlsx', index_col=0)
```

```python
df
```

```python
# сортируем по переменной "Экономическая дискриминация"
df = df.sort_values(by='Экономическая дискриминация', ascending = True)
```

```python
# график
my_color =['darkturquoise', 'orangered','goldenrod', 'orchid', 'green', 'purple']
df.plot.barh(x='Рейтинг', width = 0.9, color = my_color, stacked = True);
```

```python
my_color =['gold', 'tomato', 'crimson', 'lightseagreen', 'seagreen', 'darkorchid']
df.plot.barh(x='Рейтинг', width = 0.9, color = my_color, figsize=(6,8))
plt.xlabel('Количество показателей');
```
